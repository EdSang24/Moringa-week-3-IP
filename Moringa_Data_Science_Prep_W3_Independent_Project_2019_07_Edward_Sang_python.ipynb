```python
# importing the pandas library

import pandas as pd
```

**Loading dataset**

```python
# Importing csv file

df1 = pd.read_csv('/content/Telcom_dataset.csv', header=1, names=('PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID'))
df1
```

```python
# Importing csv file

df2 = pd.read_csv('/content/Telcom_dataset2.csv',  header=1, names=('PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID'))
df2
```

```python
# importing dataset

df3 = pd.read_csv('/content/Telcom_dataset3.csv',  header=1, names=('PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID'))
df3
```

```python
# importing dataset

df4 = pd.read_csv('/content/cells_geo.csv', delimiter=';', header=1, names=('CITY','STATUS','LOCALISATION','DECOUPZONE','ZONENAME','LONGITUDE','LATITUDE','REGION','CELL_ID','SITE_ID'))
df4
```

**Merging the datasets**

```python
#combine the datasets into one dataset

df_merged = pd.concat((df1, df2, df3), axis = 0)
df_merged
```

```python
#changing the date time column to date time format

df_merged['DATE_TIME'] = pd.to_datetime(df_merged['DATE_TIME'])
df_merged
```

```python
# dropping columns where all the values are missing

df_merged = df_merged.dropna(axis = 1, how = 'all')
df_merged
```

```python
# dropping rows where all the values are missing

df_merged_cleaned = df_merged.dropna(how = 'all')
df_merged
```

```python
# filling missing data with 0

df_merged.fillna(0)
```

```python
# merging our merged datasets with cells_geo.csv dataset
# name it df_data

df_data = df_merged.merge(df4, how='left', on='CELL_ID')
df_data
```

```python
# finding the most used product using mean

Product_use= df_data['VALUE'].groupby(df_data['PRODUCT'])
Product_use.mean()
```

```python
# finding most used cities using mean and grouping them by their CELL_IDs

Cell_ID_use= df_data['VALUE'].groupby(df_data['CELL_ID'])
Cell_ID_use.mean()
```
